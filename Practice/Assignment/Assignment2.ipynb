```python
import matplotlib.pyplot as plt
import numpy as np
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1 / (1 + math.exp(-item)))
    return a

#Plot sigmoid function - Target
x = np.arange(-10, 10, .1)
sig = sigmoid(x)
plt.plot(x, sig)
#Hypothesis Points
hx = [-10, -5.0, 0.0, 5.0, 10.0]
hy = [0.3, 0.8, 0.0, 0.5, 0.2]
plt.plot(hx, hy, 'ro')
plt.ylabel('Sigmoid Function')
plt.show()
```

```python
def sigmoidn(x):
        return (1 / (1 + math.exp(-x)))

#Create hypothesis graph
plt.plot(hx, hy, '-o')
#Target Function
plt.plot(x, sig)
#Graph anchors of sigmoid function
anchorx = [-7.5, -5.0, 2.5, 7.5]
anchory = [sigmoidn(anchorx[0]), sigmoidn(anchorx[1]), sigmoidn(anchorx[2]), sigmoidn(anchorx[3])]
print('Anchor X: ', anchorx)
print('Anchor Y: ', anchory)
plt.plot(anchorx, anchory, 'rx')
plt.show()
```

```python
def cost(anchorx):
    sum = 0
    for i in anchorx:
        if i >= -10 and i <= -5:
            y = (0.1 * i) + 1.3
            yt = sigmoidn(i)
            yf = yt - y
            sum += (yf * yf)
        if i > -5 and i <= 0:
            y = 0.16 * i
            yt = sigmoidn(i)
            yf = yt - y
            sum += (yf * yf)
        if i > 0 and i <= 5:
            y = 0.10 * i
            yt = sigmoidn(i)
            yf = yt - y
            sum += (yf * yf)
        if i > 5 and i <= 10:
            y = -0.06 * i + 0.8
            yt = sigmoidn(i)
            yf = yt - y
            sum += (yf * yf)
    return sum / len(anchorx)
    
weight = [(0.1 * -7.5) + 1.3, (0.16 * 5.0) + 0, (0.10 * 2.5), (-0.06 * 7.5) + 0.8]
cost(anchorx)
print(weight)
```

```python
def gradient(w, i, rate):
    yi = sigmoidn(i)
    newy = (yi - w)
    return newy * rate

check = [True, True, True, True]
rate = 0.1

for i in range(0, 100):
    #Old hypothesis graph
    plt.plot(hx, hy, '-o')
    #New hypothesis graph
    plt.plot(anchorx,weight, '-o')
    #Target Function
    plt.plot(x, sig)
    #Graph anchors of sigmoid function
    plt.plot(anchorx, anchory, 'rx')
    plt.show()
    for j in anchorx:
        if j >= -10 and j < -5:
            if check[0] == True:
                check[0] = False
                new = gradient(weight[0], j, rate)
                weight[0] += new
            else:
                new = gradient(weight[0], j, rate)
                weight[0] += new
        if j >= -5 and j <= 0:
            if check[1] == True:
                check[1] = False
                new = gradient(weight[1], j, rate)
                weight[1] += new
            else:
                new = gradient(weight[1], j, rate)
                weight[1] += new
        if j > 0 and j <= 5:
            if check[2] == True:
                check[2] = False
                new = gradient(weight[2], j, rate)
                weight[2] += new
            else:
                new = gradient(weight[2], j, rate)
                weight[2] += new
        if j > 5 and j <= 10:
            if check[3] == True:
                check[3] = False
                new = gradient(weight[3], j, rate)
                weight[3] += new
            else:
                new = gradient(weight[3], j, rate)
                weight[3] += new
                

        
```
